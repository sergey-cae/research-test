```python
import cot_reports as cot
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Fetch COT data
data = cot.cot_year(year=2024, cot_report_type='disaggregated_fut')
```

```python
st_year = 2010
end_year = 2024
data_frames = []

for i in range(st_year, end_year + 1):
    single_year = pd.DataFrame(cot.cot_year(i, cot_report_type='disaggregated_fut'))
    data_frames.append(single_year)  # Add each year's DataFrame to the list

# Combine all DataFrames using pd.concat()
data = pd.concat(data_frames, ignore_index=True)
```

```python
# NET VALUES

# Importing required libraries
import matplotlib.pyplot as plt
from matplotlib import colormaps
import numpy as np

# Data Preparation
df = data.copy()
df["Report_Date_as_YYYY-MM-DD"] = pd.to_datetime(df["Report_Date_as_YYYY-MM-DD"])

# Filter for WTI crude oil
crude_oil_markets = [
    "WTI-PHYSICAL - NEW YORK MERCANTILE EXCHANGE",
]
df_crude = df[df["Market_and_Exchange_Names"].isin(crude_oil_markets)].copy()
df_crude.sort_values("Report_Date_as_YYYY-MM-DD", inplace=True)


# Convert trader category columns to integers if necessary
df_crude[f"M_Money_Positions_Long_All"] = df_crude[f"M_Money_Positions_Long_All"].astype(int)
df_crude[f"Traders_M_Money_Long_All"] = df_crude[f"Traders_M_Money_Long_All"].astype(int)
df_crude[f"M_Money_Positions_Short_All"] = df_crude[f"M_Money_Positions_Short_All"].astype(int)
df_crude[f"Traders_M_Money_Short_All"] = df_crude[f"Traders_M_Money_Short_All"].astype(int)


# Calculate Net OI and Net Traders
df_crude["Net_OI"] = df_crude["M_Money_Positions_Long_All"] - df_crude["M_Money_Positions_Short_All"]
df_crude["Net_Traders"] = df_crude["Traders_M_Money_Long_All"] - df_crude["Traders_M_Money_Short_All"]

# Extract year and calculate medians and quartiles
df_crude["Year"] = df_crude["Report_Date_as_YYYY-MM-DD"].dt.year
median_net_traders = df_crude["Net_Traders"].median()
q1_net_traders = df_crude["Net_Traders"].quantile(0.25)
q3_net_traders = df_crude["Net_Traders"].quantile(0.75)

median_net_oi = df_crude["Net_OI"].median()
q1_net_oi = df_crude["Net_OI"].quantile(0.25)
q3_net_oi = df_crude["Net_OI"].quantile(0.75)

# Prepare plot
plt.figure(figsize=(14, 10))
colors = colormaps["tab20"].resampled(len(df_crude["Year"].unique()))

# Add grey bands for medians and quartiles
plt.axhspan(q1_net_oi, q3_net_oi, color='gray', alpha=0.2, label="Median Quartiles")
plt.axvspan(q1_net_traders, q3_net_traders, color='gray', alpha=0.2)

# Plot data by year
unique_years = df_crude["Year"].unique()
for i, year in enumerate(unique_years):
    year_data = df_crude[df_crude["Year"] == year]
    plt.scatter(
        year_data["Net_Traders"],
        year_data["Net_OI"],
        label=f"{year} Net",
        color=colors(i),
        alpha=0.6
    )

# Highlight the most recent week
recent_row = df_crude.iloc[-1]
plt.scatter(
    recent_row["Net_Traders"],
    recent_row["Net_OI"],
    color="black",
    edgecolor="black",
    s=200,
    # label="Most Recent Net",
    alpha=0.8
)

# Add median lines
plt.axhline(median_net_oi, color='black', linestyle='--', linewidth=1, label="Median Net OI")
plt.axvline(median_net_traders, color='black', linestyle='--', linewidth=1, label="Median Net Traders")

# Labels and Title
plt.title("DP Indicator for Crude Oil (Net OI and Net Traders, with Medians and Quartiles)", fontsize=16)
plt.xlabel("MM Net Number of Traders", fontsize=14)
plt.ylabel("MM Net OI", fontsize=14)
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.legend(loc="center left", bbox_to_anchor=(1, 0.97), fontsize=10)
plt.grid(alpha=0.3)

# Customize x and y-axis ticks
plt.xticks(
    ticks=range(int(df_crude["Net_Traders"].min()), int(df_crude["Net_Traders"].max()) + 10, 10),
    labels=[str(x) for x in range(int(df_crude["Net_Traders"].min()), int(df_crude["Net_Traders"].max()) + 10, 10)],
    rotation=0
)

plt.yticks(
    ticks=range(int(df_crude["Net_OI"].min()), int(df_crude["Net_OI"].max()) + 50000, 50000),
    labels=[f"{y//1000}K" for y in range(int(df_crude["Net_OI"].min()), int(df_crude["Net_OI"].max()) + 50000, 50000)]
)

# Set limits
plt.ylim(df_crude["Net_OI"].min() - 10000, None)
plt.tight_layout()
plt.show()


```

```python
df = data.copy()

# Filter for WTI crude oil
crude_oil_markets = [
    "WTI-PHYSICAL - NEW YORK MERCANTILE EXCHANGE",
    # "WTI FINANCIAL CRUDE OIL - NEW YORK MERCANTILE EXCHANGE",
    # "CRUDE OIL, LIGHT SWEET-WTI - ICE FUTURES EUROPE",
]

df_crude = df[df["Market_and_Exchange_Names"].isin(crude_oil_markets)].copy()

# df_crude["M_Money_Positions_Short_All"] = -df_crude["M_Money_Positions_Short_All"]


# Prepare data for plotting
df_crude["Report_Date_as_YYYY-MM-DD"] = pd.to_datetime(df_crude["Report_Date_as_YYYY-MM-DD"])
df_crude.sort_values("Report_Date_as_YYYY-MM-DD", inplace=True)

# Combine all trader categories for crude oil and prepare data for visualization
categories = [
    ("Prod_Merc", ["blue", "lightblue"], "Producers & Merchants"),
    # ("Swap", ["orange", "peachpuff"], "Swap Dealers"),
    ("Swap", ["red", "lightcoral"], "Swap Dealers"),
    ("M_Money", ["green", "lightgreen"], "Managed Money"),
    # ("Other_Rept", ["red", "lightcoral"], "Other Reportables"),
    ("Other_Rept", ["gray", "lightgray"], "Other Reportables"),
]


# Convert trader category columns to integers if necessary
for category, _, _ in categories:
    df_crude[f"Traders_{category}_Long_All"] = df_crude[f"Traders_{category}_Long_All"].astype(int)
    df_crude[f"Traders_{category}_Short_All"] = df_crude[f"Traders_{category}_Short_All"].astype(int)

# Calculate concentration for each category
for category, _, _ in categories:
    # Long concentration
    df_crude[f"{category}_Concentration_Long"] = (
        df_crude[f"{category}_Positions_Long_All"] / df_crude["Open_Interest_All"]
    ) * 100

    # Short concentration
    try:
        df_crude[f"{category}_Concentration_Short"] = (
            df_crude[f"{category}_Positions_Short_All"] / df_crude["Open_Interest_All"]
        ) * 100
    except:
         df_crude[f"{category}_Concentration_Short"] = (
            df_crude[f"{category}__Positions_Short_All"] / df_crude["Open_Interest_All"]
        ) * 100       

# Prepare data for plotting
plt.figure(figsize=(14, 10))

for category, color, label in categories:
    # Long positions
    plt.scatter(
        df_crude[f"Traders_{category}_Long_All"],
        df_crude[f"{category}_Concentration_Long"],  # Concentration for long positions
        label=f"{label} Long",
        color=color[0],
        alpha=0.6
    )
    # Short positions
    plt.scatter(
        df_crude[f"Traders_{category}_Short_All"],
        -df_crude[f"{category}_Concentration_Short"],  # Concentration for short positions
        label=f"{label} Short",
        color=color[1],
        alpha=0.6,
        edgecolors="black"
    )

# Highlight the most recent week
recent_row = df_crude.iloc[-1]
for category, color, label in categories:
    plt.scatter(
        recent_row[f"Traders_{category}_Long_All"],
        recent_row[f"{category}_Concentration_Long"],
        color=color[0],
        edgecolor="black",
        s=200,
        alpha=0.8
    )
    plt.scatter(
        recent_row[f"Traders_{category}_Short_All"],
        -recent_row[f"{category}_Concentration_Short"],
        color=color[1],
        edgecolor="black",
        s=200,
        alpha=0.8
    )

# Labels and title
plt.title("Trader Group Concentration for Crude Oil", fontsize=16)
plt.xlabel("Number of Traders", fontsize=14)
plt.ylabel("Long and Short Concentration", fontsize=14)
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.legend(loc="center left", bbox_to_anchor=(1, 0.78), fontsize=10)
plt.grid(alpha=0.3)

# Customize x-axis
tick_range = range(0, max([df_crude[f"Traders_{c[0]}_Long_All"].max() for c in categories]) + 1, 10)
plt.xticks(
    ticks=tick_range,
    labels=[str(x) for x in tick_range],
    rotation=0
)

plt.show()
```
